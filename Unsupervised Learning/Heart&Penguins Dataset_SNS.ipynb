```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = 'data/heart.csv'
df = pd.read_csv(data)
df.head()
```

```python
df = df.drop(['sex', 'cp', 'fbs', 'restecg', 'exang', 'slope', 'ca', 'thal'], axis = 1)
```

```python
sns.pairplot(df, hue = 'target')
plt.show()
```
